```python
import numpy as np
from collections import defaultdict
```

```python
def new_explored_state():
    return np.full((18, ), 0.)

Q = defaultdict(new_explored_state)
```

```python
def totuple(a):
    try:
        return tuple(totuple(i) for i in a)
    except TypeError:
        return a
```

```python
episodes = 50
G = 0

alpha = 0.618

for episode in range(1,episodes+1):
    done = False
    G, reward = 0,0

    state = env.reset().reshape((100800,))
    state = tuple(state)
    print(f"Episode: {episode+1}")
    print(f"Initial state Q: {Q[state]}")
    counter = 0
    firstState = state
    done = False
    while not done:
        env.render()
        action = np.argmin(Q[state])  #1
        state2, reward, done, info = env.step(action) #2
        state2.resize((100800,))
        state2 = tuple(state2)
        reward += 1.
        Q[state][action] += alpha * (reward + np.min(Q[state2]) - Q[state][action]) #3

        counter += 1
        G += reward
        state = state2
    print("Solved in {} Steps with a total reward of {}".format(counter, G))
    print(f"Final state Q: {Q[state]}")
    env.close()    
    
```
